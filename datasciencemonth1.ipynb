```python
''' Week 2: Data Structures and Functions

#### Markdown Cell: Objectives
markdown
## Objectives
- Learn Python programming basics.
- Practice data manipulation with basic Python, NumPy, and Pandas.
- Implement basic Python scripts for data processing and visualization.


#### Markdown Cell: Theory
markdown
## Theory: Study Python Lists, Tuples, Dictionaries, Sets, Functions, Lambda Functions, Recursion, and List Comprehension
- **Lists**: Ordered, mutable sequences (e.g., `[1, 2, 3]`).
- **Tuples**: Immutable sequences (e.g., `(1, 2, 3)`).
- **Dictionaries**: Key-value pairs (e.g., `{'a': 1}`).
- **Sets**: Unordered collections of unique elements (e.g., `{1, 2, 3}`).
- **Functions**: Reusable code blocks (e.g., `def add(a, b): return a + b`).
- **Lambda**: Anonymous functions (e.g., `lambda x: x**2`).
- **Recursion**: Function calling itself.
- **List Comprehension**: Concise list creation (e.g., `[x**2 for x in range(5)]`).'''


#### Code Cell: Hands-On - Work with Data Structures and Write Functions for Data Transformations

# Sum of squares
numbers = [1, 2, 3]
squares = [x**2 for x in numbers]
print("Sum of squares:", sum(squares))

# Filter even numbers
even_nums = [x for x in numbers if x % 2 == 0]
print("Even numbers:", even_nums)


#### Code Cell: Client Project - Write a Script for Data Cleaning

# Remove duplicates, filter data
data = [1, 2, 2, 3, 3, 4, 5, 5]
unique_data = list(dict.fromkeys(data))
filtered_data = [x for x in unique_data if x > 2]
print("Filtered unique data:", filtered_data)


#### Markdown Cell: Submit - Summary
## Submit: Python Script and Summary
 #**Summary of Concepts Learned**: Lists and dictionaries manage data, functions enable reusable logic, and list comprehension simplifies transformations.
```

```python
# Week 4: Data Visualization with Matplotlib and Seaborn

# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Set the style for better visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Load the tips dataset
tips = sns.load_dataset('tips')

# Basic Matplotlib plot: Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(tips['total_bill'], tips['tip'], alpha=0.5)
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()

# Seaborn advanced plot: Pairplot
sns.pairplot(tips, hue='time')
plt.show()

# Seaborn heatmap (fix: use only numeric columns)
plt.figure(figsize=(8, 6))
numeric_tips = tips.select_dtypes(include=[np.number])
sns.heatmap(numeric_tips.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Create a figure with multiple subplots
plt.figure(figsize=(12, 8))

# Scatter plot
plt.subplot(2, 1, 1)
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', size='size')
plt.title('Scatter Plot: Total Bill vs Tip by Time')

# Histogram
plt.subplot(2, 1, 2)
sns.histplot(data=tips, x='tip', bins=20, hue='sex')
plt.title('Histogram: Tip Distribution by Sex')

plt.tight_layout()
plt.show() 

```
